```python
#import
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
import numpy as np
```

```python
#functions for scraping and cleaning
def scrape_table(table_url):
    #get table
    page = requests.get(table_url)
    page_content = page.content
    soup = BeautifulSoup(page_content, 'html.parser')
    table = soup.find_all('table', {'class': 'regular'})
    
    #read scraped data into list
    all_rows = []
    for t in table:
        rows = t.find_all('tr')
        for row in rows:
            value = row.get_text()
            all_rows.append(value)
            
    #convert list to dataframe
    df = pd.DataFrame(all_rows)
    return df

def split_table(df):
        df = df[0].str.split('\n', expand = True)
        return df
```

```python
import datetime

df_stock_dict = {}
dict_shortnames = {'Basic_Materials_Sector': 'bm', 'Communication_Services_Sector':'comm', 'Consumer_Cyclical_Sector':'conc', 'Consumer_Defensive_Sector':'cond','Energy_Sector':'energy',
         'Financial_Services_Sector': 'fin', 'Healthcare_Sector':'hc', 'Industrials_Sector':'ind', 'Real_Estate_Sector':'real', 'Technology_Sector': 'tech', 
         'Utilities_Sector': 'u'}
for i in ['Basic_Materials_Sector', 'Communication_Services_Sector', 'Consumer_Cyclical_Sector', 'Consumer_Defensive_Sector', 'Energy_Sector',
         'Financial_Services_Sector', 'Healthcare_Sector', 'Industrials_Sector', 'Real_Estate_Sector', 'Technology_Sector', 
         'Utilities_Sector']:
    df = pd.read_csv(i + '.csv')
    df = df.drop('Unnamed: 0', axis = 1)
    #df = df.set_index('Date')
    #df = df.apply(pd.to_numeric, errors='coerce')
    #df.index = pd.to_datetime(df.index)
    #df['Date'] = pd.to_datetime(df['Date'])
    df['Date'] = pd.to_datetime(df['Date'])
    df['Id'] = np.nan
    temp = datetime.datetime(2015, 8, 23)
    # temp = '2015-08-23 00:00:00'
    id = int(0)
    for j in range(len(df)):
        if df['Date'].iloc[j] > temp:
            temp = df['Date'].iloc[j]
            df['Id'].iloc[j] = id
        else:
            id +=1
            df['Id'].iloc[j] = id
            temp = df['Date'].iloc[j]
    df_group_mean = df.groupby('Id', as_index = False)['Open'].mean()
    df_group_sd = df.groupby('Id')['Open'].std()
    df = pd.merge(df,df_group_mean,on='Id',how='left', suffixes=('','_group_mean'))
    df = pd.merge(df,df_group_sd,on='Id',how='left', suffixes=('','_group_sd'))
    df['standardized_open'] = df['standardized_open'] = (df['Open'] - df['Open_group_mean'])/df['Open_group_sd']
    df_stock_dict['df_{sec}'.format(sec = dict_shortnames[i])] = df
    df.to_pickle('df_{sec}.pkl'.format(sec = dict_shortnames[i]))
df_stock_bm = pd.read_csv('Basic_Materials_Sector.csv')
```
